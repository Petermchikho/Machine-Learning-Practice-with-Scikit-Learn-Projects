```python
import opendatasets as od
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Downloading the dataset

```python
od.download('https://www.kaggle.com/datasets/akshaydattatraykhare/diabetes-dataset')
```

loading the data into a dataframe

```python
data=pd.read_csv('/content/diabetes-dataset/diabetes.csv')
```

Statistical measures

```python
data.describe()
```

0 is for non diabetic and 1 is for diabetic

```python
data.groupby('Outcome').mean()
```

separating the data

```python
X= data.drop(columns='Outcome',axis=1)
Y= data['Outcome']
```

```python
X.head(20)
```

```python
Y.head()
```

Standardizing the data to a uniform range

```python
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data=scaler.transform(X)
```

```python
standardized_data
```

```python
X=standardized_data
Y=data['Outcome']
```

Splitting the data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Initializing the model

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

```python
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)

print('Accuracy score of the trainig data : ',training_data_accuracy)
```

```python
X_train_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_train_prediction,Y_test)

print('Accuracy score of the : ',test_data_accuracy)
```

Prediction

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
